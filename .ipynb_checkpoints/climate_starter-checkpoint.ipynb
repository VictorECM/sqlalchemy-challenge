# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base=automap_base()
# reflect the tables
Base.prepare(engine,reflect=True)
```

```python
# View all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement =Base.classes.measurement
Station =Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session=Session(engine)
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Starting from the most recent data point in the database. 
for row in session.query(Measurement.date).order_by(Measurement.date.desc()).limit(1):
    lastdate = row[0]

print(lastdate)

finaldate = dt2.strptime(lastdate,'%Y-%m-%d')

print(finaldate)
# Calculate the date one year from the last date in data set.
new_day = finaldate.day
new_month = finaldate.month
new_year = finaldate.year-1

if new_day>9 and new_month>9:
    initdate = (f'{new_year}-{new_month}-{new_day}')
elif new_day<9 and new_month>9:
    initdate = (f'{new_year}-{new_month}-0{new_day}')
elif new_day>9 and new_month<9:
    initdate = (f'{new_year}-0{new_month}-{new_day}')
else:
    initdate = (f'{new_year}-0{new_month}-0{new_day}')

print(initdate) 

# Perform a query to retrieve the data and precipitation scores
for row in session.query(Measurement.date, Measurement.prcp).filter((Measurement.date<finaldate) & (Measurement.date>initdate)):
    print(row)

# Save the query results as a Pandas DataFrame and set the index to the date column
df = pd.read_sql(session.query(Measurement.date, Measurement.prcp).filter((Measurement.date<finaldate) & (Measurement.date>initdate)).statement, session.bind)

df = df.set_index('date')

# Sort the dataframe by date
df = df.sort_index()
```

# Exploratory Station Analysis

```python
# Design a query to calculate the total number stations in the dataset
station_count = session.query(Station.station).count()
print(station_count)
```

```python
# Design a query to find the most active stations (i.e. what stations have the most rows?)
# List the stations and the counts in descending order.
station_group = session.query(Measurement.station, func.count(Measurement.id)).group_by(Measurement.station).order_by(func.count(Measurement.id).desc()).all()
print(station_group)
```

```python
most_active_station = station_group[0][0]
print(most_active_station)
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.
temp_status = session.query(func.min(Measurement.tobs), func.max(Measurement.tobs), func.avg(Measurement.tobs)).filter((Measurement.station == most_active_station)).all()

print(temp_status)
```

```python
x = session.query(Measurement.tobs).filter((Measurement.date<finaldate) & (Measurement.date>initdate) & (Measurement.station == most_active_station)).all()
xxx = []
tmp = 0
for i in range(len(x)):
    tmp = x[i][0]
    xxx.append(tmp)
aaa = np.array(xxx)
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
plt.hist(x=aaa, bins=12, alpha=0.7, rwidth=0.85)

plt.grid(axis='y', alpha=0.75)
plt.ylabel('Frequency', fontsize=18)


plt.legend(['tobs'])

plt.show()
```
